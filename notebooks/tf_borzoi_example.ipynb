```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import h5py
import numpy as np
import tensorflow as tf
tf.config.experimental.enable_tensor_float_32_execution(False)
import baskerville
from baskerville import seqnn
import json
```

```python
model_file = "model0_best.h5" #select fold
params_file = 'params.json' # from Calico repo
```

```python
%%capture
#Take
with open(params_file) as params_open :
    params = json.load(params_open)
    params_model = params['model']
    params_train = params['train']
    
params_model['verbose'] = True

seqnn_model = seqnn.SeqNN(params_model)
seqnn_model.restore(model_file, 0)
```

```python
slices = [7531, 7532, 7533] # gtex blood tracks
```

```python
sequence_one_hot_wt = np.load('../wt_seq.npy')
wt_pred_across_folds_tf = seqnn_model(sequence_one_hot_wt[None, ...])[..., slices]
```
